```python
def get_luminosity(x):
    if x<60 and x>0:
        return 1
    elif x>=60 and x <120:
        return 2
    elif x>=120 and x <180:
        return 3
    elif x>=180 and x <240:
        return 4
    
```

```python
df['luminosity_Cat']=df['luminosity'].apply(get_luminosity)
```

```python
px.histogram(df['luminosity_Cat'])
```

```python
idx=df[df['luminosity_Cat']>1.5]['luminosity'].index
```

```python
df.drop(idx,axis=0,inplace=True)
```
